```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline

palette = sns.color_palette()

plt.rcParams["axes.facecolor"] = "lightgrey"
plt.rcParams["axes.edgecolor"] = "white"
plt.rcParams["axes.grid"] = True
plt.rcParams["grid.alpha"] = 1
plt.rcParams["grid.color"] = "#cccccc"
plt.rcParams["grid.linestyle"] = "--"
plt.rcParams["savefig.pad_inches"] = 0.2

import src.read_data
```

```python
df = src.read_data.get_atm_df()
df.head()
```

# Lags

```python
for i, atm in enumerate(df.columns):
    for lag in range(1, 8):

        fig, ax = plt.subplots(figsize=(8, 8), dpi=450)
        fig.suptitle(f"{atm.upper()} Lag:{lag}", fontsize=36)
        fig.patch.set_alpha(1)
        ax.set_facecolor('white')

        pd.plotting.lag_plot(df[atm], lag=lag, ax=ax)

        fig.savefig(
            f"./pics/lags/{atm}_lag_{lag}.png",
            bbox_inches="tight",
            pad_inches=1,
            transparent=False,
        )
        plt.close()
```

# Autocorrelations

```python
for i, atm in enumerate(df.columns):

    fig, ax = plt.subplots(figsize=(8, 8), dpi=450)
    fig.suptitle(f"{atm.upper()}", fontsize=36)
    fig.patch.set_alpha(1)
    ax.set_facecolor('white')

    pd.plotting.autocorrelation_plot(df[atm], ax=ax)

    fig.savefig(
        f"./pics/autocorr/{atm}.png",
        bbox_inches="tight",
        pad_inches=1,
        transparent=False,
    )
    plt.close()
```
